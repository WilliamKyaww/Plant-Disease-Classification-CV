# 02 Data Pipeline (15-Class)

This notebook validates dataloaders using frozen split CSVs generated by `src.prepare_splits`.
Primary task target column is `class_label`.


```python
from pathlib import Path

from src.datasets import create_dataloaders
from src.transforms import get_train_transform, get_val_transform
from src.utils import CSV_DIR, CLASS_NAMES

train_csv = str(Path(CSV_DIR) / "plantvillage_train.csv")
val_csv = str(Path(CSV_DIR) / "plantvillage_val.csv")
test_csv = str(Path(CSV_DIR) / "plantvillage_test.csv")

train_loader, val_loader, test_loader = create_dataloaders(
    train_csv=train_csv,
    val_csv=val_csv,
    test_csv=test_csv,
    train_transform=get_train_transform(),
    val_transform=get_val_transform(),
    label_column="class_label",
    batch_size=32,
    num_workers=0,
)

print("Batches - train/val/test:", len(train_loader), len(val_loader), len(test_loader))
```

```python
images, labels = next(iter(train_loader))
print("Batch tensor shape:", tuple(images.shape))
print("Batch labels shape:", tuple(labels.shape))

decoded = [CLASS_NAMES[int(i)] for i in labels[:8]]
print("First 8 class labels:")
for i, name in enumerate(decoded, 1):
    print(f" {i}. {name}")
```
