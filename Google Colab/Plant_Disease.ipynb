# Colab Runner (Context Notebook)

This notebook is optional context.
It auto-detects the repository root and runs `src.prepare_splits` plus a Colab smoke report from the repository root.


```python
from pathlib import Path
import os
import sys

CANDIDATES = [
    Path('/content/Plant-Disease-Detection-CV'),
    Path('/content/Plant Disease Detection CV'),
    Path('/content/drive/MyDrive/Plant-Disease-Detection-CV'),
]

repo_main = None
for candidate in CANDIDATES:
    if (candidate / 'src').is_dir():
        repo_main = candidate
        break

if repo_main is None:
    raise FileNotFoundError('Could not find repo root with src/. Update CANDIDATES for your Drive layout.')

os.chdir(repo_main)
if str(repo_main) not in sys.path:
    sys.path.insert(0, str(repo_main))

print('Using repo root:', repo_main)
```

```python
from src.prepare_splits import main as build_splits
from src.colab_smoke import run_colab_smoke, save_report

build_splits()
report = run_colab_smoke(repo_main_arg=str(repo_main))
paths = save_report(report)
print('Smoke passed:', report['passed'])
print('Smoke artifact:', paths['latest_json'])
```
